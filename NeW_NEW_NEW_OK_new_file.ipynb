table4['FILES'] = table4['FILES'].str.split(','):    
+ Разбивает значения в столбце 'FILES' DataFrame table4 по запятым, создавая списки значений для каждой строки.    
+ df_files = table4['FILES'].apply(lambda x: pd.Series(x)):
Создает новый DataFrame df_files, где каждый элемент списка из столбца 'FILES' в table4 становится отдельной строкой.    
Это позволяет "распаковать" списки в отдельные столбцы.    
+ df_files.columns = [f'FILE_{i}' for i in range(1, len(df_files.columns) + 1)]:
Присваивает новые названия столбцам в df_files, используя формат 'FILE_1', 'FILE_2' и т.д.    
+ table6 = pd.concat([table4[['OBJECTS']], df_files], axis=1):
Объединяет DataFrame table4 (только столбец 'OBJECTS') и DataFrame df_files по столбцам (axis=1), создавая новый DataFrame table6.    
Таким образом, table6 содержит столбец 'OBJECTS' из table4 и новые столбцы 'FILE_1', 'FILE_2' и т.д., созданные из разбитых значений столбца 'FILES'.

axis=1 означает, что операция будет применяться по столбцам, а не по строкам

```python
table4['FILES'] = table4['FILES'].str.split(',')
df_files = table4['FILES'].apply(lambda x: pd.Series(x))
df_files.columns = [f'FILE_{i}' for i in range(1, len(df_files.columns) + 1)]
table6 = pd.concat([table4[['OBJECTS']], df_files], axis=1)
```

```python
import numpy as np
```

+ for col in table7.columns: - этот цикл проходит по каждому столбцу в таблице table7.    
+ table7[col] = table7[col].apply(lambda x: ...) - для каждого столбца, он применяет функцию к каждому элементу в столбце.    
+ lambda x: np.nan if isinstance(x, str) and x.endswith(('.mp4', '.zip', '.7z', '.html', '.crdownload')) else x - это функция, которая проверяет каждый элемент в столбце. Если элемент является строкой (isinstance(x, str)) и заканчивается на .mp4, .zip, .7z, .html или .crdownload (x.endswith(('.mp4', '.zip', '.7z', '.html', '.crdownload'))), то он заменяет этот элемент на np.nan (пропуск данных). Если элемент не удовлетворяет этим условиям, то он остается без изменений.

```python
# Удалить расширения .mp4, .zip и .7z в столбцах
for col in table7.columns:
    table7[col] = table7[col].apply(lambda x: np.nan if isinstance(x, str) and x.endswith(('.mp4', '.zip', '.7z', '.html', '.crdownload')) else x)
```

```python
table7
```

### Удалить строки , где во всех столбцах значение NaN

```python
# Удалить строки, где в столбцах с FILE_1 по FILE_60 значение NaN
table_not_NaN = table7.dropna(subset=[f'FILE_{i}' for i in range(1, 61)], how='all')
```

```python
table_not_NaN
```

### Разбить колонку OBJECTS	по категориям

```python
table_not_NaN
```

```python
# Создать копию таблицы
table_not_NaN = table_not_NaN.copy()

# Разбить колонку 'OBJECTS' по запятым и получить список значений для каждой строки
table_not_NaN['OBJECTS'] = table_not_NaN['OBJECTS'].str.split(',')

# Перенести разбитые значения на строчки ниже, копируя по строчно значения
table_OBJECT_categ_obied = table_not_NaN.explode('OBJECTS')
```

```python
table_OBJECT_categ_obied
```

### Подсчет суммы по строчно

table_OBJECT_categ_obied.iloc[:, 1:] выбирает все строки и столбцы, начиная со второго столбца.    
count(axis=1) применяет функцию count() по строкам этого диапазона, то есть подсчитывает количество ненулевых значений в каждой строке.

```python
# Посчитать количество столбцов со значением в строке, кроме первого столбца 'OBJECTS'
table_OBJECT_categ_obied['Summa']= table_OBJECT_categ_obied.iloc[:, 1:].count(axis=1)
```

### Объединить одинаковые колонки и суммировать по столбцу Summa

```python
# объединяем повторяющиеся значения в столбце 'Столбец_1' и суммируем значения в столбце 'Столбец_2'
Obshii = table_OBJECT_categ_obied.groupby('OBJECTS')['Summa'].sum().reset_index()
```

```python
# выводим значения в порядке убывания
df_sorter = Obshii.sort_values(by='Summa', ascending=False)
# выводим результат
df_sorter
```
